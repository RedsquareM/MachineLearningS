```python
wine3 = arff.loadarff('../knn/dataset_191_wine.arff')
wine4 = pd.DataFrame(wine3[0])
```

```python
def mapping(data,feature):
    featureMap=dict()
    count=0
    for i in sorted(data[feature].unique(),reverse=True):
        featureMap[i]=count
        count=count+1
    data[feature]=data[feature].map(featureMap)
    return data
```

```python
wine5= mapping(wine4, "class")
```

```python
wine5.sample(10)
```

```python
def normalize(dataset):
    dataNorm=((dataset-dataset.min())/(dataset.max()-dataset.min()))*1
    dataNorm["class"]=dataset["class"]
    return dataNorm
```

Nearest-neighbor regression
Nearest-neighbor regression is an adaptation of nearest-neighbor algorithms to regression. In nearest-neighbor algoriths each point in the dataset determines its class by looking at the classes of a certain number points nearest to it (and perhaps aggregating their votes somehow). The extension to the regression case couldn't be simpler: instead of just taking the majority class, average the values of the contributing points (perhaps with some kind of weighing, perhaps not).

```python
def euclidean_dist(x,y):
        """
        Euclidean distance between points x and y
        """
        if type(x)==np.float64:
            return (((y-x)**2)**0.5)
        return sum((y[i]-x[i])**2 for i in range(len(x)))**0.5
```

```python
 def manhattan_dist(self,x,y):
        """
        Manhattan distance between points x and y
        """
        if type(x)==np.float64:
            return (abs(x-y))
        return sum(abs(x[i]-y[i]) for i in range(len(x)))
```

```python
def manhattan(arr,n):
    arr.sort()
    distance = 0 
    sum = 0
    for i in range(n):
        distance += (arr[i]*i - sum)
        sum += arr[i]
    return distance 
def distancesum (x,y,n):
    return manhattan(x,n)+manhattan(y,n)

x = [1,2,3]
y = [4,5,6]
n = len(x)
print(distancesum(x,y,n))
```

```python
 def chebyshev_dist(self, x, y):
        points = []
        for i, (x, y) in enumerate(zip(x, y)):
            points.append((
                x + y + i, 
                x + y - i,
                x - y + i,
                x - y - i
            ))
        return max(max(dim) - min(dim) for dim in zip(*points))
```
